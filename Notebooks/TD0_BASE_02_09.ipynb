# IMPORT DES LIBRAIRIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mpl
```

```python
r7 = pd.read_csv('./data/restaurant_1_week_007.csv')
r32 = pd.read_csv('./data/restaurant_1_week_032.csv')
```

# Fusionner les fichiers

```python
 merge_r= pd.concat([r7, r32], ignore_index=True)
 merge_r
```

# CALCUL DU CHIFFRE D AFFAIRE

```python
merge_r['CA'] = merge_r['Product Price'] * merge_r['Quantity'] 
```

# CREATION D'AUTRE COLONNE DATE, HEURE

```python
merge_r["Order Date"]=pd.to_datetime(merge_r["Order Date"])
merge_r["Date_heure"]=merge_r["Order Date"].dt.floor("h")
#r7["Date_heure"]=r7["Order Date"].dt.strftime('%Y-%m-%d-%H:00')
merge_r['day'] = merge_r['Order Date'].dt.day
merge_r['month'] = merge_r['Order Date'].dt.month
```

```python
merge_r
```

# Regrouper le chiffre d'affaire par heure et jour

```python
heure_ca=merge_r.groupby("Date_heure")["CA"].sum()
heure_ca
```

```python
heure_ca=merge_r[["Date_heure","CA"]].groupby('Date_heure').agg({"CA":"sum"})
heure_ca=heure_ca.reset_index()
```

# OBJECTIF: tracer le CA en fonction du temps (date avec heure)

```python
mpl.plot(heure_ca['Date_heure'],heure_ca['CA'], ls='--',c="red",marker='o')
mpl.xticks(rotation=45)

# Ajouter des labels et un titre pour plus de clarté
mpl.xlabel('Date et Heure')
mpl.ylabel('Chiffre d\'Affaires')
mpl.title('Chiffre d\'Affaires au Fil du Temps')
```

# SOLUTION AVEC UNE BOUCLE

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import matplotlib.dates as mdates 
import locale 
import glob  # Utilisé pour lister les fichiers dans le répertoire 

# Définir la langue locale en français 
locale.setlocale(locale.LC_TIME, 'fr_FR.UTF-8') 

# Lister tous les fichiers CSV dans le répertoire courant 
csv_files = glob.glob('./data/*.csv') 

# Liste pour stocker les DataFrames traités 
all_dataframes = [] 

# Boucle sur chaque fichier CSV 
for file in csv_files: 
    # Lire le fichier CSV 
    df = pd.read_csv(file) 
     
    # Calculer la colonne CA 
    df["CA"] = df["Quantity"] * df["Product Price"] 
     
    # Convertir 'Order Date' en datetime 
    df['Order Date'] = pd.to_datetime(df['Order Date']) 
     
    # Extraire la date arrondie à l'heure 
    df['Date'] = df['Order Date'].dt.strftime('%Y-%m-%d %H:00') 
    df['date_floor'] = df['Order Date'].dt.floor('h') 
    df['jour_commande'] = df['Order Date'].dt.day 
    df['heure_commande'] = df['Order Date'].dt.hour 
    df['mois_commande'] = df['Order Date'].dt.month 
    df['année_commande'] = df['Order Date'].dt.year 
    
    # Grouper les données par 'Date' et calculer la somme de 'CA' 
    df_grouped = df.groupby("Date").sum("CA").reset_index() 
     
    # Ajouter le DataFrame traité à la liste 
    all_dataframes.append(df_grouped) 

# Concaténer tous les DataFrames traités 
final_dataframe = pd.concat(all_dataframes, ignore_index=True) 

# Afficher les premières lignes du DataFrame final pour vérifier 
print(final_dataframe.head()) 
 
# Plot des données concaténées 
fig, ax = plt.subplots() 
ax.plot_date(final_dataframe["Date"], final_dataframe["CA"], linestyle="solid") 
plt.xticks(rotation=45)  # Rotation des dates pour une meilleure lisibilité 
plt.xlabel('Date') 
plt.ylabel('CA') 
plt.title('Chiffre d\'affaires par date') 
plt.grid(True) 
plt.show()

mpl.plot(heure_ca['Date_heure'],heure_ca['CA'], ls='--',c="red",marker='o')
mpl.xticks(rotation=45)
```
